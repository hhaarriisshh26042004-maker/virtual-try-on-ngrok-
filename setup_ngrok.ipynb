```python
!pip install gradio -q

import gradio as gr
from PIL import Image
import os

def make_dir():
    os.makedirs("/content/inputs/test/cloth", exist_ok=True)
    os.makedirs("/content/inputs/test/cloth-mask", exist_ok=True)
    os.makedirs("/content/inputs/test/image", exist_ok=True)
    os.makedirs("/content/inputs/test/image-parse", exist_ok=True)
    os.makedirs("/content/inputs/test/openpose-img", exist_ok=True)
    os.makedirs("/content/inputs/test/openpose-json", exist_ok=True)

def virtual_tryon(model_img, cloth_img):
    """Process virtual try-on"""
    make_dir()

    # Save inputs
    model_img.save("/content/inputs/test/image/model.jpg")
    cloth_img.save("/content/inputs/test/cloth/cloth.jpg")

    # Run the pipeline
    os.system("python /content/clothes-virtual-try-on/run.py")

    # Load and return output
    output_dir = "/content/output"
    if os.path.exists(output_dir):
        output_files = os.listdir(output_dir)
        if output_files:
            result = Image.open(f"{output_dir}/{output_files[0]}")
            os.system("rm -rf /content/output/")
            return result
    return None

# Create Gradio interface
demo = gr.Interface(
    fn=virtual_tryon,
    inputs=[
        gr.Image(type="pil", label="ðŸ‘¤ Person Image"),
        gr.Image(type="pil", label="ðŸ‘• Cloth Image")
    ],
    outputs=gr.Image(type="pil", label="âœ¨ Result"),
    title="Virtual Try-On",
    description="Upload a person image and a cloth image to see the virtual try-on result."
)

# Launch with public URL
demo.launch(share=True)
```

```python
from flask import Flask, request, send_file, jsonify
from pyngrok import ngrok
from pyngrok.conf import PyngrokConfig
from PIL import Image
import threading
import io
import os

app = Flask(__name__)

def make_dir():
    os.system("cd /content/ && mkdir -p inputs && cd inputs && mkdir -p test && cd test && mkdir -p cloth cloth-mask image image-parse openpose-img openpose-json")

@app.route("/")
def home():
    return jsonify("hello world")

@app.route("/api/transform", methods=['POST'])
def begin():
    make_dir()
    print("data received")
    cloth = request.files['cloth']
    model = request.files['model']

    cloth_img = Image.open(cloth.stream)
    model_img = Image.open(model.stream)

    cloth_img.save("/content/inputs/test/cloth/cloth.jpg")
    model_img.save("/content/inputs/test/image/model.jpg")

    os.system("python /content/clothes-virtual-try-on/run.py")

    op = os.listdir("/content/output")[0]
    op = Image.open(f"/content/output/{op}")
    buffer = io.BytesIO()
    op.save(buffer, 'png')
    buffer.seek(0)
    os.system("rm -rf /content/output/")
    return send_file(buffer, mimetype='image/png')

# Start ngrok tunnel FIRST
config = PyngrokConfig(ngrok_path="/usr/local/bin/ngrok")
public_url = ngrok.connect(5000, pyngrok_config=config)
print(f"\n{'='*50}")
print(f"ðŸš€ PUBLIC URL: {public_url}")
print(f"{'='*50}\n")

# Then start Flask (this blocks, so it must be last)
app.run(port=5000)
```
